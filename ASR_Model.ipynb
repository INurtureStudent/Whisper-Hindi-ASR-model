# 6. Audio Recognition




```python
import os
import speech_recognition as sr
import ffmpeg

# Initialize the recognizer object
recognizer = sr.Recognizer()

# Function to recognize speech from an audio file
def recognize_speech(audio_file):
    # Convert audio file to WAV format using ffmpeg
    wav_audio_file = audio_file.split('.')[0] + '.wav'
    ffmpeg.input(audio_file).output(wav_audio_file).run(overwrite_output=True)

    # Use the recognizer object to recognize speech from audio data
    try:
        with sr.AudioFile(wav_audio_file) as source:
            audio_data = recognizer.record(source)  # Record audio from the WAV file
            text = recognizer.recognize_google(audio_data, language='hi-IN')  # Recognize speech using Google Speech Recognition
        return text
    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand the audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"
    finally:
        # Clean up - remove the temporary WAV file
        os.remove(wav_audio_file)
```

# 7. Speech to Text

```python
import os
import speech_recognition as sr
import ffmpeg

# Initialize the recognizer object
recognizer = sr.Recognizer()

# Function to recognize speech from an audio file
def recognize_speech(audio_file):
    # Convert audio file to WAV format using ffmpeg
    wav_audio_file = audio_file.split('.')[0] + '.wav'
    ffmpeg.input(audio_file).output(wav_audio_file).run(overwrite_output=True)

    # Use the recognizer object to recognize speech from audio data
    try:
        with sr.AudioFile(wav_audio_file) as source:
            audio_data = recognizer.record(source)  # Record audio from the WAV file
            text = recognizer.recognize_google(audio_data, language='hi-IN')  # Recognize speech using Google Speech Recognition
        return text
    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand the audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"
    finally:
        # Clean up - remove the temporary WAV file
        os.remove(wav_audio_file)

# Path to the folder containing audio files
folder_path = '/content/drive/MyDrive/GV_Eval_3h/Audio'

# Path to the output text file
output_file_path = '/content/drive/MyDrive/GV_Eval_3h/ResultText.txt'

# Open the output text file for writing
with open(output_file_path, 'w') as output_file:
    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.mp3'):  # Check if the file is an MP3 file
            file_path = os.path.join(folder_path, filename)

            # Recognize speech from the audio file
            recognized_text = recognize_speech(file_path)

            # Write the recognized text to the output file
            output_file.write(f"File: {filename}\n")
            output_file.write(f"Recognized Text: {recognized_text}\n\n")
```
